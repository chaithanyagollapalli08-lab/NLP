data -> lower -> cleaning(removing urls,punctuations) -> tokenization -> stopword removal -> stemming/lemmatization


```python
data='On June 21, 2026, Sachin Tendulkar visited the Microsoft office in Hyderabad, India, to attend an Artificial Intelligence conference. During the event, he discussed Machine Learning, Deep Learning, Natural Language Processing, and Computer Vision with researchers from Google, Amazon, and Infosys. More than 5,000 participants from different countries attended the conference, and many students were excited to learn about emerging technologies such as Generative AI, Large Language Models, Retrieval-Augmented Generation, and Robotics. After the conference, Sachin shared his thoughts on social media, saying that technology will continue to transform healthcare, education, finance, and transportation in the coming years.'
```

```python
data=data.lower()
data
```

```python
import string
cleaned_data=''
for i in data:
  if i not in string.punctuation:
    cleaned_data=cleaned_data+i
print(cleaned_data)
```

```python
import nltk
nltk.download('punkt')
nltk.download('punkt_tab')
```

```python
from nltk import word_tokenize
tokens=word_tokenize(cleaned_data)
print(tokens)
```

```python
c=[]
for i in tokens:
  if not i.isdigit():
    c.append(i)
print(c)
```

```python
nltk.download('stopwords')
nltk.download('words')
```

```python
from nltk.corpus import stopwords
l=[]
for i in c:
  if i not in stopwords.words('english'):
      l.append(i)
print(l)
```

```python
from nltk.stem import PorterStemmer
stem=[]
ps=PorterStemmer()
for i in l:
  stem.append(ps.stem(i))
print(stem)
```

```python
nltk.download('wordnet')
```

```python
from nltk.stem import WordNetLemmatizer
le=[]
lem=WordNetLemmatizer()
for i in l:
  le.append(lem.lemmatize(i))
print(le)
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
obj=TfidfVectorizer()
x=obj.fit_transform(le)
print(obj.get_feature_names_out())
print(x.toarray())
```
