bag of words

```python
from sklearn.feature_extraction.text import CountVectorizer
sentences=['i love nlp','i love python']
obj=CountVectorizer()
x=obj.fit_transform(sentences)
print(x.toarray())
```

N_Grams

```python
from sklearn.feature_extraction.text import CountVectorizer
sentences=['i love nlp','i love python']
obj=CountVectorizer(ngram_range=(2,2))
x=obj.fit_transform(sentences)
print(x.toarray())
```

TF-IDF

```python
from sklearn.feature_extraction.text import TfidfVectorizer
sentences=['i love nlp','i love python']
obj=TfidfVectorizer()
x=obj.fit_transform(sentences)
print(x.toarray())
```

POS_tagging

```python
import nltk
nltk.download('punkt')
nltk.download('punkt_tab')
nltk.download('averaged_perceptron_tagger_eng')
```

```python
from nltk import word_tokenize
from nltk import pos_tag
s='i love python'
t=word_tokenize(s)
print(pos_tag(t))
```

Named Entity Recognition

```python
nltk.download('words')
nltk.download('maxent_ne_chunker_tab')
```

```python
from nltk import word_tokenize
from nltk import pos_tag
from nltk import ne_chunk
s='Sachin lives in Mumbai'
t=word_tokenize(s)
pos=pos_tag(t)
print(ne_chunk(pos))
```

```python
import spacy
d=spacy.load('en_core_web_sm')
print(d.pipe_names)
doc=d('Arjun lives in Hyderabad')
for i in doc.ents:
  print(i.text,i.label_)
```

Dependency parsing

```python
import spacy
nlp=spacy.load('en_core_web_sm')
d=nlp('ramu eats apple')
for i in d:
  print(i.text,i.dep_,i.head.text)
```

```python
from gensim.models import Word2Vec
sentences=[['i','love','nlp'],['i','love','python']]
obj=Word2Vec(sentences,vector_size=50,min_count=1)
print(obj.wv['love'])
```

```python
from gensim.models import Word2Vec
sentences=[['i','love','nlp'],['i','love','python']]
obj=Word2Vec(sentences,vector_size=50,min_count=1)
print(obj.wv.most_similar('love'))
```
